```python
#In Python we have Primitive types(numbers, strings, booleans), Complex types(range etc). This range function returns a range object which is Iterable. Strings are also iterable.
#Nested Loop
for i in range(5):
  for j in range(3):
      print(f"1st_value: {i}, 2nd_value: {j}")

for x in "Python":
  print(x)

#We have another complex type, that is LIST. Which we use to store a list of objects
for y in [1, 2, 3, 4, 5]:
   print(y)
```

```python
#Function(A parameter is the input that you define for your function, while an argument is the actual value for a given parameter)

def greet(f_name, l_name):
    print(f"Full Name is: {f_name} {l_name}")


greet("Mosh", "Hamedani")

#In Python all function returns a value "none" by default in case we doesn't specify any return statement inside the function

print(greet("Mahmud", "Ahmedinezad"))

#Keyword Parameter

def addition(val_1, val_2):
    return val_1 + val_2


print(addition(val_1=30, val_2=70))

#Default Argument(Required and Optional parameter[optional parameter MUST be placed after the required parameter])

def multiplication(val_1, val_2 =5):
    return val_1 * val_2


print(multiplication(30))
print(multiplication(30, 10)) #Optional parameter is replaced by the 2nd argument

#X number of argument

def multi(*numbers):
    print(*numbers)
    multi_result = 1
    for i in numbers:
        multi_result *= i
    return multi_result


print(multi(1, 2, 3, 4 ,5))
```

```python
#XX number of argument(When we use double asterisk, we can pass multiple key value pairs, or multiple keyword arguments to a function and PYTHON will automatically package them into a dictionary. So the user object inside the function definition is a dictionary and by using the bracket notation we can access value of various keys into the dictionary)

def user_data(**user):
    print(user)
    print(user["name"])
    print(user["age"])


user_data(id = 1, name = "John Doe", age = 44)

#FIZZ_BUZZ

def fiz_buz(value):
    if (value % 3 == 0) and (value % 5 == 0):
        return "fizzbuzz"
    elif (value % 5 == 0):
        return "buzz"
    elif (value % 3 == 0):
        return "fizz"
    else:
        return value
    
    
print(fiz_buz(29))
    
```
